```python
import pandas as pd
import numpy as np
import glob
```

```python
#get all csv files
files= glob.glob("*.csv")
files
```

```python
#combine them
df=pd.concat([pd.read_csv(file) for file in files],ignore_index=True)
df
```

```python
print('total rows:',len(df))
df.head()
```

```python
#saving combined files

df.to_csv("flipkart_mobile_data.csv",index=False)
```

cleaning price and,MRP columns

```python
#coverting column to string first

df['price']=df['price'].astype(str)
```

```python
#apply cleaning

df['price']=df['price'].astype(str)
df['price']=df['price'].str.replace('₹','')
df['price']=df['price'].str.replace(',','')
df['price']=pd.to_numeric(df['price'], errors='coerce')
```

```python
#MRP

df['MRP']=df['MRP'].astype(str)
```

```python
df['MRP']=df['MRP'].astype(str)
df['MRP']=df['MRP'].str.replace('₹','')
df['MRP']=df['MRP'].str.replace(',','')
df['MRP']=pd.to_numeric(df['MRP'], errors='coerce')
print(df['MRP'].dtype)
```
